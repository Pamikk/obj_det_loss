```python
import cv2
import numpy as np   
import matplotlib.pyplot as plt     
```

```python
def draw_bboxes(img,bboxes,color,th=1):
    img_ = img.copy()
    for bbox in bboxes:
        x,y,w,h = bbox
        pt1 = (int(x-w/2),int(y-h/2))
        pt2 = (int(x+w/2),int(y+h/2))
        img_ =cv2.rectangle(img_,pt1,pt2,color,thickness = th)
    return img_
```

```python
def rotate(src,ang,labels):
    h,w,_ = src.shape
    center =(w/2,h/2)
    mat = cv2.getRotationMatrix2D(center, ang, 1.0)
    dst = cv2.warpAffine(src,mat,(w,h))
    labels_ = labels.copy()
    xs,ys,ws,hs = labels[:,1:].T
    n = len(xs)
    cos = abs(mat[0,0])
    sin = abs(mat[0,1])
    print(mat)
    pts = np.stack([xs,ys,np.ones([n])],axis=1).T
    tpts = np.dot(mat,pts).T
    labels_[:,1] = tpts[:,0]
    labels_[:,2] = tpts[:,1]
    labels_[:,3] = (cos*ws + sin*hs)
    labels_[:,4] = (cos*hs + sin*ws)
    return dst,labels_



```

```python
img = np.zeros([128,96,3])
labels = np.array([[0,20,30,15,20],[0,64,32,10,15]])
print(labels.shape)
img = draw_bboxes(img,labels[:,1:],(255,0,0),-1)
plt.figure(figsize=(3,3))
plt.imshow(img)
for i,ang in enumerate(range(30,181,30)):
    dst,bboxes = rotate(img,ang,labels)
    dst = draw_bboxes(dst,bboxes[:,1:],(0,255,255))
    print(ang,bboxes[0])
    plt.figure(figsize=(3,3))
    plt.imshow(dst)

```
